```python
cat_columns = ['status', 'sku', 'category_name_1', 'sales_commission_code', 'Customer Since']
for col in cat_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df['category_name_1'].replace(['\\N', 'Others'], 'Others', inplace=True)
```

```python
df['payment_method'].unique()
```

```python
creditcard = ['Payaxis', 'bankalfalah', 'ublcreditcard', 'mcblite', 'internetbanking']
jazzcash = ['jazzwallet', 'jazzvoucher']
easypaisa= ['Easypay', 'Easypay_MA', 'easypay_voucher']
cod= ['cod', 'cashatdoorstep']
Others = ['apg', 'customercredit', 'mygateway','productcredit', 'marketingexpense', 'financesettlement']

# Replace old_values with new_values in the specified column
df['payment_method'].replace(creditcard, 'Card Payment', inplace=True)
df['payment_method'].replace(jazzcash, 'JazzCash', inplace=True)
df['payment_method'].replace(easypaisa, 'EasyPaisa', inplace=True)
df['payment_method'].replace(cod, 'COD', inplace=True)
df['payment_method'].replace(Others, 'Others', inplace=True)
```

```python
df['payment_method'].unique()
```

```python

df['status'].unique()
```

```python
df['status'].replace([
   'paid', 'cod', 'received', 'complete'], 'completed', inplace=True)


df['status'].replace([
    'canceled', 'closed'], 'canceled', inplace=True)


df['status'].replace([
    'order_refunded', 'refunde'], 'refund', inplace=True)


df['status'].replace([
    'payment_review', 'fraud', 'pending',
    'holded', 'processing', 'fraud', 'pending_paypal',
    'exchange', '\\N'], 'others', inplace=True)
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['grand_total'].resample('M').sum().dropna())
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_percentage_error
```

```python
monthly_sales = df['grand_total'].resample('M').sum().dropna()
train = monthly_sales[:-12]
test = monthly_sales[-12:]
model = ARIMA(train, order=(5, 1, 0))
model_fit = model.fit()
predictions = model_fit.forecast(steps=12)
```

```python

if len(train) >= 12:
    sma_predictions = train.rolling(window=12).mean().iloc[-12:]
else:
    raise ValueError("Insufficient data to compute SMA with a window size of 12.")
```

```python


aligned_sma_predictions = sma_predictions[-len(test):]

```

```python


aligned_sma_predictions = aligned_sma_predictions.fillna(method='bfill')

```

```python
sma_mape = mean_absolute_percentage_error(test, aligned_sma_predictions)
print(f'SMA MAPE: {sma_mape}')
```
